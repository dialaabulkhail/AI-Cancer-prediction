```python
import pandas as pd
data = pd.read_csv("cancer.csv")
```

- x attributes represent all dependent variables. (all features on the dataset).
- y attribute is used to predict whether the Tumor Cancer is Malignant or Benign.

```python
x = data.drop(columns=["diagnosis(1=m, 0=b)"])
```

```python
y = data["diagnosis(1=m, 0=b)"] 
```

- Splitting Data between training and testing sets is helpful because sometimes mapping the data can result in overfitting, which means data will be fitted nicely until you provide external data, thats where it falls down wih fitting it.

for avoiding this, we test part of the data to implement it to the first time on the algorithm so we can detect its behavior with new data

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

- sklearn is a very popular machine learning library, we will be using it to split data into training and testing sets.

- test size = 0.2 means that 20% of data is set up for testingn and the 80% left is for training.

[Scikit-learn](https://scikit-learn.org/stable/)

```python
import tensorflow as tf
model = tf.keras.models.Sequential(0)
```

- Imported Tensorflow's keras which is another popular library for artificial intellegance applications, It was used to build a neural network.

[Tensorflow's keras](https://www.tensorflow.org/api_docs/python/tf/keras)

```python
model.add(tf.keras.layers.Dense(200, input_shape=x_train.shape, activation='sigmoid'))
model.add(tf.keras.layers.Dense(200, input_shape=x_train.shape, activation="sigmoid"))
model.add(tf.keras.layers.Dense(1, input_shape=x_train.shape, activation= "sigmoid"))
```

Adding layers to the neural network
- the first one is for the x input from every feature.
- the second one is a hidden layer that x attributes go through.
- the last one is the output of either (Malignant) or (Benign).

**200 is a random unit** --> the output will be 200 neurons

Activation function is set to (Sigmoid) which can reduce model complexity and increase its accuracy --> all values are plotted between 0 and 1

[Keras Activations](https://keras.io/api/layers/activations/)

```python
model.compile(optimizer = "adam", loss = "binary_crossentropy", metrics = ["accuracy"])
```

- Compiling the model using an optimizer (adam).
- (binary_crossentropy) can be used when dealing with categorical data [Binary_crossentropy](https://www.tensorflow.org/api_docs/python/tf/keras/metrics/binary_crossentropy)
- (accuracy) was used as metrics to classify as much Tumors accurately. [Accuracy](https://www.tensorflow.org/api_docs/python/tf/keras/metrics/Accuracy)

```python
model.fit(x_train, y_train, epochs = 500)
```

- fitting training data model
- epochs/epics are number of times the algorithm is iterating over data.

**its best for the algorithm in order to learn to iterate multiple times over data, so its better to give an epic higher than the number of data.**

```python
model.evaluate(x_test, y_test)
```

- Testig the algorithm using x_test and y_test to get the accuracy, loss of the algorithm.

## Result
This AI model can predict Cancer diagnosis of Tumor with 92% accuracy
